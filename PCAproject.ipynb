```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('houseprice.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.iloc[:,1].isna().sum()
```

```python
house = data.copy()
```

```python
house.columns
```

```python
col = []
for i in range(len(data.columns)):
    if house.iloc[:,i].isna().sum()*100/len(house) >=20:
        col.append(data.columns[i])
      
    
```

```python
col
```

```python
house.drop(col,axis=1,inplace=True)
```

```python
house.head()
```

```python
house.info()
```

```python
house.set_index('Id',inplace=True)
```

```python
house.head()
```

```python
from sklearn.impute import SimpleImputer

for i in house:
    if(house[i].isnull().values.any()):
        
        if((house[i].dtype == 'float64') or (house[i].dtype == 'int64')):
            floatimputer = SimpleImputer(missing_values = np.nan,strategy = 'mean')
            house[i] = floatimputer.fit_transform(house[[i]])
        
        elif(house[i].dtype == 'object'):
            objectimputer = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
            house[i] = objectimputer.fit_transform(house[[i]]) 
    else:
        continue
```

```python
house.info()
```

```python
cat =[]
cont = []
for i in house.columns:
    if house.loc[:,i].dtype =='object':
        cat.append(i)
    else:
        cont.append(i)
```

```python
cat,cont
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in cat:
    house.loc[:,i] = le.fit_transform(house.loc[:,i])
```
